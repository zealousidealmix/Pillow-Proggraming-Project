```python
# Imports
from PIL.ImageDraw import Image, Draw

# Methods

#   - Setting Methods
def setupHeight():
    print("What would you like the height to be? (default 500)")
    height = get_int_input("Height", 500)
    return height
def setupWidth():
    print("What would you like the width to be? (default 500)")
    width = get_int_input("Width", 500)
    return width
def setupColor():
    print("How strong should Red be? (default and minimum 0, maximum 255)")
    red= get_int_input("Red", 0, 0, 255)
    print("How strong should Blue be? (default and minimum 0, maximum 255)")
    blue= get_int_input("Blue", 0, 0, 255)
    print("How strong should Green be? (default and minimum 0, maximum 255)")
    green= get_int_input("Green", 0, 0, 255)
    return red, blue, green
def inputPosition(string_of_interest, height, width):
    print(f"Where would you like the {string_of_interest} to go?")
    print(f"What would you like the X of the {string_of_interest} go? (default and minimum 0, maximum {width})")
    x_position = get_int_input("X", 0 , 0 , width)
    print(f"What would you like the Y of the {string_of_interest} go? (default and minimum 0, maximum {height})")
    y_position = get_int_input("Y", 0 , 0 ,height)
    return [x_position, y_position]
def inputRadius():
    print("How big would you like the radius to be? (default 0)")
    radius = get_int_input("Radius", 0)
    return radius

#   - Logic Methods
def get_int_input(prompt, default, minimum=None, maximum=None):
    value = input(prompt)
    if value.isdigit():
        value = int(value)
        if (minimum is None or value >= minimum) and (maximum is None or value <= maximum):
            return value
    print("Invalid answer, setting to ", default)
    return default


def Main():
    print("HELLO!!! :D")
    print("What size would you like the image to be?")

    #   - Canvas setup
    im = Image.new("RGB", [setupWidth(), setupHeight()], setupColor())
    draw = Draw(im)
    im.show()
    #   - Drawing loop
    while True:
        print("Looking good,")
        print("what would you like to add?")

        match input("a 'point', a 'line', a 'a rectangle' or a 'circle'? (leave blank for nothing and 'exit' to leave the code)"):
            case "point", "Point":
                draw.point(inputPosition("point", im.width, im.height),
                           setupColor())
                print("Drew a point")

            case "line", "Line":
                draw.line((inputPosition("beginning of the line", im.height, im.width),
                           inputPosition("end of the line", im.height, im.width)),
                           setupColor())
                print("Drew a line")

            case "rectangle", "Rectangle", "square", "Square":
                draw.rectangle((inputPosition("left corner", im.height, im.width),
                                inputPosition("right corner", im.height, im.width)),
                                setupColor())
                print("Drew a rectangle")

            case "circle", "Circle":
                draw.circle(inputPosition("center", im.height, im.width), inputRadius() , setupColor())

            case _:
                print("Invalid answer or exit called, shutting down.")
                break


        im.show()


    print("Thank you for using this program! Bye!") # end of code


Main()
```
