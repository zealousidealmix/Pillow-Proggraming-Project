```python
import os
from PIL.ImageDraw import Image, Draw
#region Methods
#   - Setting Methods
def setup_height():
    height = get_int_input("What would you like the height to be? (default 500)","Height", 500)
    return height
def setup_width():
    width = get_int_input("What would you like the width to be? (default 500)", "Width", 500)
    return width
def setup_color(does_contain_alpha=False):
    red= get_int_input("How strong should Red be? (default and minimum 0, maximum 255)", "Red", 0, 0, 255)
    blue= get_int_input("How strong should Blue be? (default and minimum 0, maximum 255)", "Blue", 0, 0, 255)
    green= get_int_input("How strong should Green be? (default and minimum 0, maximum 255)", "Green", 0, 0, 255)
    if does_contain_alpha:
        alpha = get_int_input("How strong should Alpha be? (default and minimum 0, maximum 255)", "Alpha", 0)
        return red, green, blue, alpha
    return red, blue, green
def input_position(string_of_interest, height, width):
    print(f"Where would you like the {string_of_interest} to go?")
    x_position = get_int_input(f"What would you like the X of the {string_of_interest} go? (default and minimum 0, maximum {width})", "X", 0 , 0 , width)
    y_position = get_int_input(f"What would you like the Y of the {string_of_interest} go? (default and minimum 0, maximum {height})", "Y", 0 , 0 ,height)
    return [x_position, y_position]
def input_radius():
    radius = get_int_input("How big would you like the radius to be? (default 0)", "Radius", 0)
    return radius

#   - Input Methods
def get_safe_number(prompt):
    number = input(prompt)
    if number.isdigit():
        return int(number)
    print("Invalid answer, please try again.")
    get_safe_number(prompt)
def get_int_input(asking_prompt, prompt, default, minimum=None, maximum=None):
    print(asking_prompt)
    value = input(prompt)
    if value.isdigit():
        value = int(value)
        if (minimum is None or value >= minimum) and (maximum is None or value <= maximum):
            return value
    print("Invalid answer, setting to ", default)
    return default
#endregion

def main_step3():
    #region Image Identification and Selection
    folder = "Images"

    images = [f for f in os.listdir(folder) if f.lower().endswith((".jpg", ".jpeg", ".png", ".webp", ".AVIF"))]
    images.sort()
    if not images:
        print("No images found")
        exit(0)

    for i, name in enumerate(images, start=1):
        print(f"{i}. {name}")
    print("00, empty canvas.")
    choice = get_safe_number("Which image would you like to open?")
    if choice != 00:
        print(f"Opening {images[choice - 1]}.")
        im = Image.open(os.path.join(folder, images[choice - 1]))
    else:
        print("Invalid answer or empty canvas chosen, opening empty canvas.")
        im = Image.new("RGB", [setup_width(), setup_height()], setup_color())
    im.show()
    #endregion

    #region The drawing Cycle
    draw = ImageDraw.Draw(im)
    while True:
        #region Massive print wall
        print("'open'      ->   Open image")
        print("'point'     ->   Place singular pixel")
        print("'circle'    ->   Place Circle")
        print("'rectangle' ->   Place Rectangle")
        print("'exit'      ->   Exits program")
        #endregion
        match input("Choose option:"):
            case "open":
                im.show()
            case "point":
                draw.point(input_position("point", im.width, im.height))
            case "line":
                draw.line((input_position("beginning of the line", im.height, im.width),
                           input_position("end of the line", im.height, im.width)),
                          setup_color(), get_int_input("line width", 1, 1, im.width))
            case "rectangle":
                draw.rectangle((input_position("left corner", im.height, im.width),
                                input_position("right corner", im.height, im.width)),
                               setup_color())
            case "circle":
                draw.circle(input_position("center", im.height, im.width),
                            input_radius(), setup_color())
            case "exit":
                print("Exiting program.")
                break
    #endregion

    #region Watermark Applier
    if input("do you want to add a watermark? [y/n]") == "y":
        folder = "Watermarks"
        water_marks = [f for f in os.listdir(folder) if f.lower().endswith((".jpg", ".jpeg", ".png", ".webp", ".AVIF"))]
        if not water_marks:
            print("No watermarks found")
            exit(0)
        for i, name in enumerate(water_marks, start=1):
            print(f"{i}. {name}")
        choice = get_safe_number("Which watermark would you like to add?")
        print(f"Applying {water_marks[choice - 1]}.")
        overlay = Image.open(os.path.join(folder, water_marks[choice - 1]))
        im.paste(overlay, input_position("Watermark", im.height, im.width))
    #endregion


    if input("do you want to save image? [y/n]") == "y": # asks to save the image, if yes, then asks its name, and saves the image in the folder.
        im.save("Images/" + input("How do you want to name the image?") + ".png")
    print("Thank you for using this program! Bye!")
main_step3()
```

### Problém s `im.show()` v Colabu
`im.show` funguje jenom v **NORMALNÍM EDITORU**, takže, mám svou methodu `display_image_in_colab`, která funguje.
Je to kvůli tomu, že google disk / colab nemá .tmp file (nebo buffer), což znamená že to nejde otevřít normálně.

```python
import io
from IPython.display import Image as DisplayImage

def display_image_in_colab(pil_image):
     with io.BytesIO() as buffer:
        pil_image.save(buffer, format="PNG")
        display(DisplayImage(data=buffer.getvalue()))
```
