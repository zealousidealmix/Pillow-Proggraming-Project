```python
# Imports
import PIL
from PIL.ImageDraw import ImageDraw
# Methods

#   - Setting Methods
def setupHeight():
    height = input("Height: ")
    if height.isdigit():
        height = int(height)
        print("Setting height to ", height)
    else:
        height = 500
        print("Invalid answer, setting to 500.")
    return height
def setupWidth():
    width = input("Width: ")
    if width.isdigit():
        width = int(width)
        print("Setting width to ", width)
    else:
        width = 500
        print("Invalid answer, setting to 500.")
    return width
def setupColor():
    print("How strong should Red be?")
    red = input("Red: ")
    if red.isdigit() and 255 >= int(red) > 0:
        red = int(red)
        print("Setting Red to ", red)
    else:
        red = 0
        print("Invalid answer, setting Blue to 0.")
    print("How strong should Blue be?")
    blue = input("Blue: ")
    if blue.isdigit() and 255 >= int(blue) > 0:
        blue = int(blue)
        print("Setting Blue to ", blue)
    else:
        blue = 0
        print("Invalid answer, setting Blue to 0.")
    print("How strong should Green be?")
    green = input("Green: ")
    if green.isdigit() and 255 >= int(green) > 0:
        green = int(green)
        print("Setting Green to ", green)
    else:
        green = 0
        print("Invalid answer, setting Green to 0.")

    return red, blue, green


# Main
print("HELLO!!! :D")
print("What size would you like the image to be?")

im = PIL.Image.new("RGB", [setupHeight(), setupWidth()], setupColor())
draw = PIL.ImageDraw.Draw(im)
im.show()

print("Looking good, just to add a few more touches.")
print("What would you like to add?")
```
